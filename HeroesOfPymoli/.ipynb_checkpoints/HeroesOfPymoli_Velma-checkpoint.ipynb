```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculate number of Unique Items
number_items = len(purchase_data["Item ID"].unique())
#Calculate Average Price of unique items
avg_price = purchase_data["Price"].mean()
#Calculate Number of Purchases
number_purchases = purchase_data["Purchase ID"].count()
#Calculate Total Revenue
total_revenue = purchase_data["Price"].sum()
```

```python
type(avg_price)
```

```python
#Display summary df
summary_df = pd.DataFrame({"Number of Unique Items": [number_items],
                          "Average Price": [avg_price],
                          "Number of Purchases": [number_purchases],
                          "Total Revenue": [total_revenue],})

summary_df1 = summary_df.style.format({'Average Price': '${:,.2f}', 'Total Revenue':'${:,.2f}'})
summary_df1
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#bins for ages
print(purchase_data["Age"].max())
print(purchase_data["Age"].min())
```

```python
uniqueSN_Age = purchase_data.drop_duplicates(['SN'])
uniqueSN_Age = uniqueSN_Age.loc[:,["Age"]]
```

```python
#Establish Bin
bins= [0, 10, 15, 20, 25, 30, 35, 40, 100]
print(len(bins))
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
#Categorize the existing players using the age bins. Hint: use pd.cut()
uniqueSN_Age["Age Group"] = pd.cut(uniqueSN_Age["Age"], bins, labels=group_labels, right=False)
uniqueSN_Age.head()
```

```python
age_group = uniqueSN_Age.groupby("Age Group").count()
age_group
```

```python
age_group = age_group.rename(columns={'Age':'Total Count'})
age_group['Percentage of Players']= (age_group['Total Count']/age_group['Total Count'].sum()).mul(100).round(2)
age_group_df = pd.DataFrame(age_group)
#Set the 'Gender' as index
age_group_df
```
